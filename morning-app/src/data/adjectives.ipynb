```python
import nltk 
nltk.download('stopwords')
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize, sent_tokenize 
stop_words = set(stopwords.words('english')) 
```

```python
with open('positive.txt', 'r') as file:
    data = file.readlines()
```

```python
with open('negative.txt', 'r') as file2:
    data2 = file2.readlines()
```

```python
data
data[:] = [line.strip('\n') for line in data]
datastr = ""
for item in data:
    datastr = datastr + item + " "
```

```python
data2
data2[:] = [line.strip('\n') for line in data2]
data2str = ""
for item in data2:
    data2str = data2str + item + " "
```

```python
data2str
```

```python
datastr
```

```python
nltk.download('punkt')
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
tokenized = sent_tokenize(datastr) 
for i in tokenized: 
      
    # Word tokenizers is used to find the words  
    # and punctuation in a string 
    wordsList = nltk.word_tokenize(i) 
  
    # removing stop words from wordList 
    wordsList = [w for w in wordsList if not w in stop_words]  
  
    #  Using a Tagger. Which is part-of-speech  
    # tagger or POS-tagger.  
    tagged = nltk.pos_tag(wordsList) 
  
    print(tagged)
```

```python
tokenized = sent_tokenize(data2str) 
for i in tokenized: 
      
    # Word tokenizers is used to find the words  
    # and punctuation in a string 
    wordsList = nltk.word_tokenize(i) 
  
    # removing stop words from wordList 
    wordsList = [w for w in wordsList if not w in stop_words]  
  
    #  Using a Tagger. Which is part-of-speech  
    # tagger or POS-tagger.  
    tagged2 = nltk.pos_tag(wordsList) 
  
    print(tagged2)
```

```python
posadj = []
```

```python
negadj = []
```

```python
for word, form in tagged:
    if form == 'JJ':
        posadj.append(word)
```

```python
for word, form in tagged2:
    if form == 'JJ':
        negadj.append(word)
```

```python
negadj
```

```python
posadj = [word.capitalize() for word in posadj if len(word) <= 10]
negadj = [word.capitalize() for word in negadj if len(word) <= 10]
```

```python
words = {
    'positive': posadj,
    'negative': negadj
}
```
